```python
# hasznaltauto demo

from selenium import webdriver
import time

driver = webdriver.Chrome('/home/mihaly/R_codes/web_scraping/class_5/chromedriver')
driver.get("https://www.hasznaltauto.hu/")
print('sleep')
time.sleep(10)
driver.find_element_by_css_selector("#mainSearch > div > div:nth-child(4) > div:nth-child(8) > div > div > button > span.total-count.visible-xs-inline-block.visible-sm-inline-block.visible-md-inline-block.visible-lg-block").click()
time.sleep(10)

for i in range(1,5):
    driver.find_element_by_css_selector(".haicon-uj-jnyil-kicsi").click()
    page = driver.page_source
    my_file = open('data_html/page_'+str(i)+'.html', 'w')
    my_file.write(page)
    my_file.close()
    
    time.sleep(10)
driver.close()
```

```python
# mnb
from selenium import webdriver
import time
import pandas as pd
atmnevek = []
atmcimek =[]
idk = []

driver = webdriver.Chrome('/home/mihaly/R_codes/web_scraping/class_5/chromedriver')
driver.get("https://fiokesatmkereso.mnb.hu/web/index.html")
time.sleep(20)




for i in range(1,30):
    nevek =driver.find_elements_by_css_selector("div.list-item-data-title")
    cimek = driver.find_elements_by_css_selector(".list-item-data-title+ .list-item-data-details")
    m_idk = driver.find_elements_by_css_selector(".list-item")
    
    hosz = len(cimek)
    for i in range(0,len(cimek)):
        atmnevek.append(nevek[i].text)
        atmcimek.append(cimek[i].text)
        idk.append(m_idk[i].get_attribute('id'))
        

    driver.find_element_by_css_selector("#pager-control-next img").click()
    time.sleep(1)
    
driver.quit()

df = pd.DataFrame({'cimek':atmcimek,'atmnevek':atmnevek, 'id':idk})

df.to_csv('all_data_text.csv')
df

```
